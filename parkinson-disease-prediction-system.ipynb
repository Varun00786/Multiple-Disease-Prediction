```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("parkinsons.csv")
df
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
df.isnull().sum()
```

```python
features=df.drop(columns=["name","status"])
labels=df.status
```

```python
features
```

```python
labels
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(features,labels,test_size=0.3,stratify=labels)
train_x
```

```python
test_x
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
np.mean(cross_val_score(KNeighborsClassifier(algorithm='brute'),features,labels,cv=6))
```

```python
np.mean(cross_val_score(DecisionTreeClassifier(),features,labels,cv=6))
```

```python
np.mean(cross_val_score(SVC(kernel='linear'),features,labels,cv=6)) # {'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'}
```

```python
np.mean(cross_val_score(RandomForestClassifier(),features,labels,cv=6))
```

```python
np.mean(cross_val_score(RandomForestClassifier(n_estimators=40,criterion="log_loss"),features,labels,cv=6)) 
# criterion : {"gini", "entropy", "log_loss"}, default="gini"
```

```python
np.mean(cross_val_score(LogisticRegression(max_iter=1000),features,labels,cv=6)) 
```

```python
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(train_x,train_y)
```

```python
model.predict([[122.4,148.65,113.819,0.00968,0.00008,0.00465,0.00696,0.01394,0.06134,0.626,0.03134,0.04518,0.04368,0.09403,0.01929,19.085,0.458359,0.819521,-4.075192,0.33559,2.486855,0.368674]])
```
