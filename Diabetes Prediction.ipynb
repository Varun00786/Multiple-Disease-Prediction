```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
df=pd.read_csv("Healthcare-Diabetes.csv")
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df["Outcome"].value_counts()
```

```python
features=df.iloc[:,1:-1]
labels=df.iloc[:,[-1]]
display(features,labels)
```

```python
train_x,test_x,train_y,test_y=train_test_split(features,labels,stratify=labels,test_size=0.2)
display(train_x,train_y,test_x,test_y)
```

```python
cross_val_score(DecisionTreeClassifier(criterion="entropy"),train_x,train_y,cv=5)
```

```python
cross_val_score(KNeighborsClassifier(n_neighbors=7),train_x,train_y,cv=5)
```

```python
cross_val_score(SVC(kernel= "poly"),train_x,train_y,cv=5)
```

```python
cross_val_score(RandomForestClassifier(),train_x,train_y,cv=5)
```

```python
model=KNeighborsClassifier(n_neighbors=7)
```

```python
model.fit(train_x,train_y)
```

```python
model.score(test_x,test_y)
```

```python
model.predict([[3,187,70,22,200,36.4,0.408,36]])
```
